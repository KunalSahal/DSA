## Methods

```python
# Original string will not be changed 
print(str.upper()) # str.lower
print(str)

# SWAPCASE
print("smallCAPITAL", "-> swapcase() ->", "smallCAPITAL".swapcase())
```

#### 1. rstrip()

```python
# Remove trailing character(from the end)
# Not from the beginning 
print(str.rstrip("y"))
print(str.rstrip("one")) # no change as not in trial
```

#### 2. replace()

```python
# will replace as many sp characters present 
sample_1 = "Silver Spoon"
print(sample_1.replace("Sp", "M"))
print(sample_1)
```

#### 3. split()

```python
# String -> list as per the character metion in the ("")
# can be ',' or ' ' etc 
sample_2 = "Hi this is string that is going to convert in string "
print(sample_2.split(' '))
```

#### 4. capitalize()

```python
# To capatalize the first word and fix the unnecessary capital letters
# Full stop is not recognized
sample_3 = "hi mAy namE is kUnal. using capitalize method"
print(sample_3.capitalize())
```

#### 5. title()

```python
# Every words first letter is capitalized
print(sample_3.title())
```

#### Extra : Best way to manage word formatting

```python
import re

random = "hi mAy namE is kUnal. using replacement technique"

result = re.sub(r'(^|[.]\s*)([a-zA-Z])',
                lambda m: m.group(1) + m.group(2).upper(),
                random.lower())

print(result)
```

#### 6. center()

```python
# pushes string towards center by adding spaces 
# length of the string increases because of the spaces added
# how for in the center depends on the value given 
# "Hello".center(width)
print("Hello".center(20))

# To make automatic center adjustment as per the window size 
import shutil

width = shutil.get_terminal_size().columns
print("Hello".center(width))
```

#### 7. count()

```python
# to find how many times it is repeating
print(sample_2)
print("No.of times i used =",sample_2.count("i")) 
```

#### 8. endswith() & startswith()

```python
# checks weather ends with the given character or not 
# returns boolean only
# even space can be used to check 
print(sample_2.endswith("!"))
print(sample_2.endswith("g "))

# checking with slicing 
# end value is inclusive 
test = sample_2.endswith("ing", 2, 17) # endswith("char", start,end)
print(test)

print("\nstartswith()")
# same goes for startswith but from beginning
print(sample_2.startswith("hi")) # casesensitive so false  
print(sample_2.startswith(" ", 2, 8)) # because 2nd index is space 
```

#### 9. find() & index()

```python
# both return index value of the first letter of the first occurence 
test = "hi this is this is"
print(test.find("is"), "==", test.index("is"))
# considers all the characters mentioned but returns only the first letter's index
# '-1' if value not found
print(test.find("hi!"))

# index will provide error message instead of a value return. This is the only difference between these two
print(test.index('hi!'))
```

#### 10. isalnum() & isalpha()

```python
# isalnum -> Trues if contains A-Z | a-z | 0-9 else false
# isalnum -> Flase if space is there 
print(("test "+"00").isalnum(), "because space is there")
print("-----------------------------------------------------")
# isalpha -> True if A-Z | a-z else false 
print(("test"+"as").isalpha(), "-> only alphabetical values")
print(("test "+"as").isalpha(), "-> space")
print(("test"+"00").isalpha(), "-> numercial values are there")
```

#### 11. islower() & isupper()

```python
# True if all characters are lowercase else false 
print("fghj fdf".islower()) # all lowercase 
print("AShhjjjjjjj".islower()) # some uppercase 
print("--------")
# True if all characters are uppercase else false
print("AShhjjjjjjj".isupper()) # some uppercase 
print("ASDFGHJ".isupper()) # some uppercase 
```

#### 12. isprintable()

```python
# True if all elements are printable else false 
print("ghj hj".isprintable())
print("ghj hj \n".isprintable()) # \n prints next line and is not printable
```

#### 13. istitle()

```python
# True if every word's first letter is capital else false 
print("as Asd hgIhj".istitle())
print("As This Is First Caps".istitle())
```
