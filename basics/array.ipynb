# ARRAYS / LIST

```python
# syntax -> listName[default=0 : default=len(listName)]
ls = [1, 4, 3, 343, 6435, 2, 8, 35, 543, 7]
print(type(ls))
print(ls[-2])
print(ls[len(ls)-2]) # better way to write negative idexing -> converted in positive index
```

### Checking Value's presence

```python
# if value in ls -> used to check weather a value exsist in the list or not 
# int value and string value will not be considered same 
if 4 in ls:
    print(True)
else:
    print(False)
if "4" in ls:
    print(True)
else:
    print(False)
```

### Slicing

```python
# Slicing -> listName[start : end : jumpindex]
print(ls[:])
print(ls[2:5])
print("---------------------------------------------")
# Slicing does not affect main list. Returns new list
print(ls)
var = ls[2:5]
print(var)
print("---------------------------------------------")
# Now using jump index
print(ls)
print(ls[::3])
print("---------------------------------------------")
```

### List Comprehension
    Syntax = [ <expression_if_true> if <condition> else <expression_if_false> for item in iterable ]

```python
# ADDING VALUES VIA STATEMENT
wholeNumbers = [i for i in range(0, 10)]
print(wholeNumbers)
```

```python
# USING IF-ELSE
evenNumbers = [i if i%2==0 else "odd" for i in wholeNumbers ]
print(evenNumbers)
```

```python
# USING IF-ELIF-ELSE
val = [
    "multiple of two" if i%2==0
    else "multiple of three" if i%3==0
    else i 
    for i in wholeNumbers
]
print(val)
```

```python
# STRING EXAMPLE 
# Only if used after for 
names = ["john", "don", "motu", "patlu", "to"]
namesWith_o = [
    i for i in names
    if "o" in i
]
print(namesWith_o)
```

```python
# STRING FILTERATION BY LENGTH 
namesWithLength_3 = [
    i if len(i)<=3
    else "len>3"
    for i in names
]
print(namesWithLength_3)
```

### METHODS

```python
# every Method will change the main list as it is mutable
ls.reverse()
print(ls)
ls.sort(reverse=True) # By default ascending
print(ls)
```

```python
# to check index value of any particular value 
ls.index(343)
```

```python
# to count repetation of any value
# NOTE : in jupyter print is not required but elsewhere print will be required to catch the value 
ls.count(7)
```

```python
# To create list without referencing 
m = ls.copy()
print(m)
# any changes made to m will not affect to ls
m.append(000)
print("printing m value =",m)
print("printing ls value =",ls)
# below will reference original one so affect the original value as well 
ref = ls
ref.append(0)
print("printing value changed due to referencing",ls)
```

```python
# To insert value on a particular location 
# Value of the mentioned location will succed with one as per position 
m.insert(4, "newValue")
print(m)
```

```python
# to insert whole new list in a exsisting list = APPEND
# to insert just the values = EXTEND  
new1 = ['@', '$']
m.append(new1)
print(m)
m.extend(new1)
```

```python
# CONCATENATE LIST
# when adding new list values without changing exsisting list 
new2 = ls + new1
print(new2)
```
