## METHODS

#### 1. union() & update()
1. New set will contain all the values from different sets without repetation 
2. No.of sets can be any 



```python
val1 = {1, 2, 3, 4, 5,}
val2 = {3, 4, 7}
val4 = {9, 8, 7}
val3 = val1.union(val2, val4) # return new string 
print(val1)
print(val2)
print(val3)
```

```python
# update() changes the main set itself with the union method
val1.update(val2)
print(val1)
```

#### 2. intersection() & intersection_update()
1. New set contains all the same values



```python
val3 = val1.intersection(val2)
print(val3)
```

```python
# to change the main set itself 
val2.intersection_update(val3)
print(val2)
```

#### 3. symmetric_difference() & symmetric_difference_update()
1. new set will contain all the values that are not common
2. symmetric_difference_update() -> to change value in the chosen set 



```python
val3.update((7, 8, 9)) # can't just add 7, 8, 9 they shoud be in some form of structure
print(val1)
print(val3)
val4 = val1.symmetric_difference(val3) # {3, 4} is removed as it is common in both 
print(val4)
```

```python
print("before", val2)
val2.symmetric_difference_update(val4)
print(f"after {val2}") # fstring
```

#### 4. difference() & difference_update()

new set contains elements of A remaining after subtracting elements of B



```python
print(val1)
print(val2)
print(val1.difference(val2)) # empty so set() returned
print(val1, "-", val3, "=", val1.difference(val3)) # returns new set 
```

```python
# to change the original set 
print(val1)
val1.difference_update(val3)
print(val1)
```

#### 5. isdisjoint() BOOLEAN
1. True if no same element else false
2. opposite of intersection

```python
temp = {7,8}
print(val1, "is dijoint of", val2, "->", val1.isdisjoint(val2))
print(val1, "is dijoint of", temp, "->", val1.isdisjoint(temp))
```

#### 6. issuperset() BOOLEAN
    
1. syntax : A.issuperset(B) 
2. True if B is derived from A else false 

```python
print(val2, "is superset of", val1, "->", val2.issuperset(val1))
print(val1, "is superset of", temp, "->", val1.issuperset(temp))
```

#### 7. issubset() BOOLEAN
    
1. syntax : A.issubset(B) 
2. True if A is derived from B else false 
3. opposite of issuperset()

```python
print(val1, "is subset of", val2, "->", val1.issubset(val2))
print(val1, "is subset of", temp, "->", val1.issuperset(temp))
```

#### 9. remove() & discard()
1. Both are used to remove a specific item 
2. error show if value not present in case of remove(). discard() will not show any error and runs the next lines
3. remove() better than discard() for error handling 

```python
val1.remove("NewValue")
print(val1)
val1.discard(0)
print("printed eventhough discard has an error")
val1.remove(0)
print(val1) # can't be printed as error encountered above 
```

#### 10. pop()

2. can delete any item as set is an unordered structure

#### 11. clear() & del 
1. del -> delete entire set -> <b>Returns NameError</b> 
2. clear -> empties the set returns set() displaying now its empty
3. both work with <b>list</b> 

```python
# clear()
temp = {1, 2, 4}
temp.clear()
print(temp)
temp90 = [1, 2, 4]
temp90.clear()
print(temp90)

# del keyword
del temp
# print(temp)
del temp90
print(temp90)
```
