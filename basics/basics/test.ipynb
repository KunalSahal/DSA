```python
var = [2, 8, 6, 7]
target = 9
dict = {}
result = []

for index, value in enumerate(var):
    diff = target - value
    if diff in dict:
        result = dict[diff] + index
    dict[value] = index
   
if result == []:
    print("Value not found")
else:
    print(result)
```

```python
var = [-12, -11, 4, -6, 8, 9, -10]
left, right = 0, len(var)-1

while left < right:
    
    if (var[left] < 0 and var[left+1] > 0) or (var[left] > 0 and var[left+1] < 0):
        var[left] = var[left] + var[left+1]
        var[left+1] = var[left] - var[left+1]
        var[left] = var[left] - var[left+1]
    
    if var[right] < 0 and var[right-1] > 0:
        var[right] = var[right] + var[right-1]
        var[right-1] = var[right] - var[right-1]
        var[right] = var[right] - var[right-1]
        
    left += 1
    right -= 1
        
print(var)
```
