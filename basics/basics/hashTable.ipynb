<h1><b>Hashing Concept</b></h1>
This concept means generating fix size output using hash function 

<ol>
  <li><b>Hash Function</b><br>
    value -> unicode -> sum all there unicode -> % with 10 to get index value between 0-9 known as <b>Hash Code</b>
  </li>
  <br>
  <li><b>Hash Table</b>
    <ul>
      <li>Array containing values placed as per index returned by the <b>Hash Function</b></li>
      <li>If multiple values with same index known as <b>Collision</b></li>
      <li>Collision is solved by <b>Chaining</b> where values in stored in the same index in a bucked which is multidimensional list or <b>Linked List</b></b></li>
    </ul>
  </li>
</ol>
<h3>Advantages</h3>
<ol>
  <li>Fastest Lookup(searching):</li>
  O(1) -> index is known so directy reaching to the values is possible
  <li>Efficient for Large Data:</li>
  Great for frequency counting, indexing, caching
  <li>No Need for Sorted Data</li>
  <li>Avoids Linear Search</li>
</ol>
<h3>Disadvantages</h3>
<ol>
  <li>Collisions:</li>
  Two keys can map to the same index. Collisions slow down performance
  <li>No sorted order</li>
</ol>



```python
# Direct integer have no Unicode -> either use them directly or convert them into string 
# String values 0-9 have unicode values 
# isinstance(object, classinfo) -> to check specified data type 

def hashFunction(value):
    sum_of_chars = 0
    if isinstance(value, int):
        value = str(value)
        
    for val in value:
        sum_of_chars += ord(val)
        
    return sum_of_chars%10

ls = [None]*10

def hashTable(value):
    hashCode = hashFunction(value)
    ls[hashCode] = value

hashTable(10)
hashTable(11)
print(ls)
```
