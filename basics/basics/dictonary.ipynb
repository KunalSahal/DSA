## METHODS

### 1. get()

```python
val = {
    1:"kunal", 
    2:"tanu", 
    3:"pritam"
}
print(val) # for complete dictonary
print(val.get(5)) # throws NONE if value not found  
print(val[5]) # single key value -> throws key error if key not found 
```

### 2. keys()
1. returns all the keys of your dictonary
2. Used to iterate dictonary via keys 

```python
print(val.keys())

for keys in val.keys():
    print(keys, "->", val.get(keys))
```

### 3. values()
1. returns all the values in the dict

```python
print(val.values())

for values in val.values():
    print(values)
```

### 4.items()
returns (key, value) in the form of <b>TUPLES</b>

```python
print(val.items())

for i in val.items():
    print(i)
```

```python
# to get keys and values seperately 
for key, val in val.items():
    print(f"{key} -> {val}")
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
    
node_1 = Node(10)
node_2 = Node(20)

node_1.next = node_2
print(node_2.next)
```
