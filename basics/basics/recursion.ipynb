#### Factorial 

```python
# fact(5) = 5 * 4 * 3 * 2 * 1
#         = 5 * fact(4)
# fact(n) = n * fact(n-1)
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(5))
```

#### Fibonacci  

```python
# fibonacci(5) = 0 + 1 + 1 + 2 + 3 
def fibonacci(val):
    try:
        float(val)
        if val == 0:
            return 0
        if val == 1:
            return 1
        return fibonacci(val-1) + fibonacci(val-2)
    except(ValueError):
        return ("invaild input")
    
print(fibonacci(6))
```
