# <p style="text-align: center;">Two Pointer Algorithm</p>
1. Using two variables or pointers pointing towards different index values moving them either towards each other , away from each other or parallel to each other 
2. Used for linear data structures such as array(list), string or linked list 
3. Use case when input data follows a predictable pattern ex<b> sorted list, palindrome, pair of values satisfing a condition (like two sum) or result generated using two values</b>

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        start = 0
        end = len(s)-1
        while (start != len(s)-1) :
            if s[start]<"A" or (s[start]>"Z" and s[start]<"a") or s[start]>"z" :
                if  not (47<ord(s[start])<58) : 
                    start+=1
                    continue 
            if s[end]<"A" or (s[end]>"Z" and s[end]<"a") or s[end]>"z" :
                if  not (47<ord(s[end])<58) : 
                    end-=1
                    continue   
            if s[start] == s[end] or (abs(ord(s[start])-ord(s[end]))==32):
                if (ord(s[start])>57 and ord(s[end])>57) or (ord(s[start])<57 and ord(s[end])<57) or s[start]==s[end] : 
                    start+=1
                    end-=1
                else :
                    print(start,"=",s[start],end,"=",s[end])
                    return False
            else : 
                return False
        return True # 37 ms & 5.11% beats
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        start = 0
        end = len(s)-1
        while (start < end) :
            if not s[start].isalnum():
                start+=1
                continue 
            if not s[end].isalnum():
                end-=1
                continue   
            if s[start].lower()==s[end].lower():
                    start+=1
                    end-=1
            else : 
                return False
        return True 
```
